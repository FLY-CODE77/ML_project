# Background 
- Using Infrared image predict CWSI(Crop Water Stress INDEX)
- Lack of moisture caus poor crop production and poor quality 
- CWSI is stress index value is between 0 ~ 1
- temp is one of the main factor(CWSI)
- When Leaf temperature is lower than Air temperature (1 ~ 4 C), CWSI become 0 and crop reproduce well 
- when Air temperature is higher than Leaf temperature (1 ~ 4 C), CWSI beacome 1 and crop stop reproduce
